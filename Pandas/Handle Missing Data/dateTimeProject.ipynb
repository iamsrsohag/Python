**Simple Project**
 1. Read the data from CSV file.
 2. Change the index to date.
 3. Change the N/A value according to Time method.
 4. Make a date range and re-index the range.
 5. Save the CSV file.

```python
#1. Read the data from CSV file.

import pandas as pd
df = pd.read_csv("weather_data.csv")
```

```python
 #2. Change the index to date.
    
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index("day",inplace=True)
df
```

```python
#3. Change the N/A value according to Time method.
time_df = df.interpolate(method="time")
time_df
```

```python
#4. Make a date range and re-index the range.

range_df = pd.date_range("2017-01-01","2017-01-15")
index = pd.DatetimeIndex(range_df)
df = df.reindex(index)
df.fillna(method="ffill") 
df
df.interpolate(method="time")
```

```python
# 5. Save the CSV file.
df.to_excel("dateTimeProject.xlsx",sheet_name='DateTime')
```
