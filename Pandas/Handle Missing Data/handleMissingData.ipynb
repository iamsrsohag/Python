```python
import pandas as pd
df = pd.read_csv("weather_data.csv")
df
```

```python
#Converting Day Columns as Date.

df = pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index("day",inplace=True)
df
```

```python
#To Fill the NaN Values.

new_df = df.fillna("ReplaceValue")
new_df
```

```python
#To Fill the NaN Values for specific Columns.

new_df = df.fillna({
    "temperature": 0,
    "windspeed":"Ojana Batash",
    "event":"Ordinary"
})
new_df
```

```python
#Replace NaN value with Previous Data
df = pd.read_csv("weather_data.csv")
cp_df = df.fillna(method="ffill") 
cp_df
```

```python
#Replace NaN value with Next Data

df = pd.read_csv("weather_data.csv")
cp_df = df.fillna(method="bfill") 
cp_df
```

```python
#Replace NaN value with Next Column

df = pd.read_csv("weather_data.csv")
cp_df = df.fillna(method="bfill",axis="columns") 
cp_df
```

```python
#Replace NaN value with Limiting how many values will be filled.

df = pd.read_csv("weather_data.csv")
cp_df = df.fillna(method="ffill",limit=1) #Replace Only 1 empty value.
cp_df
```

**Linear Interpolate (replace with middle value) for more check this link**
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.interpolate.html


```python
df = pd.read_csv("weather_data.csv")
in_df = df.interpolate(method='quadratic') #If there's no method declared it'll be linear by default.
in_df
```

```python
#Remove Rows with N/A Values. It drops if at least 1 N/A value available.

df = pd.read_csv("weather_data.csv")
na_df = df.dropna()
na_df
```

**Drop rows if it doesn't have given number of N/A values**
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html

```python
df = pd.read_csv("weather_data.csv")
na_df = df.dropna(thresh=4)
na_df
```
